## Question 2 (10 Points)
Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
# You code here (Please add comments in the code):

# My text data
data = ["The Botswana prime minister announced new regulations.",
        "Russia still not ready to leave Ukraine in piece",
        "England's Arsenal football team won the world championship.",
        "The vice president calls Obama a friend.",
        "A new AI software was released by Google.",
        "Taylor Swift release a new song."]

# Extracting features
def get_features(texts):
    from sklearn.feature_extraction.text import TfidfVectorizer
    import spacy
    import pandas as pd # Importing the pandas library
    from textblob import TextBlob
    nlp = spacy.load('en_core_web_sm')
#Vectorizer and fit initialization
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(texts)

    features = []
    for doc in texts:
        doc_features = {}
        doc_features['word_count'] = len(doc.split())
        pos_tags = [token.pos_ for token in nlp(doc)]
        doc_features['pos_tags'] = pos_tags
        entities = [(ent.text, ent.label_) for ent in nlp(doc).ents]
        doc_features['ner'] = entities
        doc_features['sentiment'] = TextBlob(doc).sentiment.polarity
        features.append(doc_features)

    return pd.DataFrame(features)#Returning features as a data frame

features_df = get_features(data)
print(features_df)#Displaying features

```

## Question 3 (10 points):
Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)."

Select the most important features you extracted above, rank the features based on their importance in the descending order.

```python
# You code here (Please add comments in the code):
#Importing important packages
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_extraction.text import TfidfVectorizer

# My text data
data = ["The Botswana prime minister announced new regulations.",
        "Russia still not ready to leave Ukraine in piece",
        "England's Arsenal football team won the world championship.",
        "The vice president calls Obama a friend.",
        "A new AI software was released by Google.",
        "Taylor Swift release a new song."]

# Assigning labels for my data ( 0 = Politics, 1 = Sports, 2 = Technology, 3 = Showbiz)
y = [0,0,1,0,2,3]

# Initializing and fitting the TF-IDF vectorizer on the text
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(data)

# Working on feature selection via mutual information
mutual_info = mutual_info_classif(X, y)

# Sorting the features based on their weight
feature_ranking = sorted(zip(mutual_info, vectorizer.get_feature_names_out()), reverse=True)

# Printing the ranked features
print("Feature Ranking (Mutual Information):")
for score, feature in feature_ranking:
    print(f"Feature: {feature}, Score: {score}")

```

## Question 4 (10 points):
Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
# You code here (Please add comments in the code):
# Installing the transformers library since am  running in a new environment)
!pip install transformers

from transformers import BertTokenizer, BertModel
import torch
from sklearn.metrics.pairwise import cosine_similarity

# Loading the BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# My text data
data = ["The Botswana prime minister announced new regulations.",
        "Russia still not ready to leave Ukraine in piece",
        "England's Arsenal football team won the world championship.",
        "The vice president calls Obama a friend.",
        "A new AI software was released by Google.",
        "Taylor Swift release a new song."]

#a query for matching
query = "Government introduces new policies for the economy."

# get_bert_embedding function to get BERT embeddings for a given text
def get_bert_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True)
    outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).detach().numpy()

# Getting the BERT embedding for the query
query_embedding = get_bert_embedding(query)

# calculating cosine similarities. This is between the query and each text
similarities = []
for text in data:
    text_embedding = get_bert_embedding(text)
    similarity = cosine_similarity(query_embedding, text_embedding)[0][0]
    similarities.append((text, similarity))

# Placing the texts in descending order
similarities.sort(key=lambda x: x[1], reverse=True)

# Printing the ranked texts based on similarity to the query
print("Arranged Texts Based on Similarity: ")
for text, score in similarities:
    print(f"Text: {text} <|> Similarity Score: {score}")




```
